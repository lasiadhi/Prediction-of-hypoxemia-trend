```python
import pandas as pd
import numpy as np
import os
from tableone import TableOne

pd.options.display.max_columns = 200
pd.options.display.max_rows = 200
```

```python
FILENAME_X_train = "../Data/data_with_baseline_PF/X_train.csv"
FILENAME_X_test  = "../Data/data_with_baseline_PF/X_test.csv"
FILENAME_y_train = "../Data/data_with_baseline_PF/y_train.csv"
FILENAME_y_test  = "../Data/data_with_baseline_PF/y_test.csv"

X_train = pd.read_csv(FILENAME_X_train, index_col=0)
X_test = pd.read_csv(FILENAME_X_test, index_col=0)
y_train = pd.read_csv(FILENAME_y_train, index_col=0).iloc[:,0]
y_test = pd.read_csv(FILENAME_y_test, index_col=0).iloc[:,0]

# merge all
X_df = pd.concat([X_train, X_test])
y_df = pd.concat([y_train, y_test])
df_features = X_df.join(y_df)
```

```python
columns = list(df_features.columns)
groupby = 'is_worse'
categorical = ['gender','ethnicity','unittype','unitstaytype','BMI_category','is_low_urine','vasopressor','min_t_gcs','max_t_gcs']
nonnormal = []
```

```python
mytable = TableOne(df_features.reset_index(drop=True),columns, categorical, groupby, nonnormal, pval=True)
```
