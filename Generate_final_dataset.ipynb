```python
df_all_labs = pd.merge(df_lab, df_lab_other, how='inner', on=['id','bin'])
```

```python
# Select required labs for the final dataset:
```

```python
lab_list = ['BUN','calcium','creatinine','glucose','bicarbonate','hematocrit','hemoglobin','INR',\
            'platelets','potassium','sodium','wbc','PT','RBC','anion gap','chloride']
feature_list = ['min_t_','max_t_','avg_t_']
final_lab_list = []
for lab_i in lab_list:
    final_lab_list = final_lab_list + [x + lab_i for x in feature_list]
```

```python
final_lab_list
```

```python
labs_to_model = df_all_labs[['id', 'bin'] + final_lab_list]
```

```python
labs_to_model.shape
```

```python
# generate lab features:
```

```python
labs_to_model_min = labs_to_model[['id'] +['min_t_' + x for x in lab_list]].groupby('id').min()
```

```python
labs_to_model_max = labs_to_model[['id'] +['max_t_' + x for x in lab_list]].groupby('id').max()
```

```python
labs_to_model_mean = labs_to_model[['id'] +['avg_t_' + x for x in lab_list]].groupby('id').mean().round(3)
```

```python
lab_features = pd.concat([labs_to_model_min, labs_to_model_max, labs_to_model_mean], axis = 1)
```

```python
lab_features.head()
```

```python
lab_features.isnull().sum()
```

```python
# Impute missing values by normal lab values 
# (OPTIONAL)
```

```python
# BUN
lab_features[[x + 'BUN' for x in feature_list]].describe([0.01,0.05,0.25,0.5,0.75,0.95,0.99])
```

```python
lab_features[[x + 'BUN' for x in feature_list]] = lab_features[[x + 'BUN' for x in feature_list]].fillna(20)
```

```python
# calcium
lab_features[[x + 'calcium' for x in feature_list]].describe([0.01,0.05,0.25,0.5,0.75,0.95,0.99])
```

```python
lab_features[[x + 'calcium' for x in feature_list]] = lab_features[[x + 'calcium' for x in feature_list]].fillna(8.6)
```

```python
# creatinine
lab_features[[x + 'creatinine' for x in feature_list]].describe([0.01,0.05,0.25,0.5,0.75,0.95,0.99])
```

# Read urine output data for last 24hr from IV start:

```python
df_uo = pd.read_csv("../Processed_data/uo_24h_before_IV.csv", index_col=False)
```

```python
df_uo.head()
```

```python
# generate features:
uo_features = df_uo.groupby('patientunitstayid')['urineoutput'].aggregate({'avg_t_urine': 'mean'})
```

```python
# Oliguria is considered to be a urinary output of less than 400 milliliters/ 24hr
uo_features['is_low_urine'] = uo_features['avg_t_urine']<400
```

```python
uo_features['is_low_urine'] = uo_features['is_low_urine']*1
```

```python
uo_features = uo_features.reset_index()
```

```python
uo_features
```

### Append outcome variable
